**Plotting Pandas**

```python
normals=pd.Series(np.random.normal(size=30))
normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
variables=pd.DataFrame({'normal':np.random.normal(size=10),
                        'gamma':np.random.gamma(1,size=10),
                        'poisson':np.random.poisson(size=10)})
print(variables)
variables.cumsum(0).plot()
```

```python
variables.cumsum(0).plot(secondary_y='normal')
```

```python
variables.cumsum(0).plot(subplots=True) #cumsum is cumulated sum
```

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(nrows=3,ncols=1,figsize=(4,12))
for i,var in enumerate(['normal','gamma','poisson']):
    variables[var].cumsum(0).plot(ax=axes[i],title=var)
axes[0].set_ylabel('cumulatiev sum')
```

**Simple Line Plots**

```python
fig=plt.figure()
ax=plt.axes()
s=np.linspace(0,20,100)
print(s)
ax.plot(s,np.sin(s));
```

```python
plt.plot(s,np.sin(s))
```

**Line Colors and Styles**

```python
plt.plot(s,np.sin(s-0),linestyle='-')
plt.plot(s,np.sin(s-1),linestyle=':')
plt.plot(s,np.sin(s-2),linestyle='-.')
```

**Adjusting the Plot:Axes Limits**

```python
plt.plot(s,np.sin(s-0))
plt.xlim(-1,15)
plt.ylim(-2,2)
```

**Labeling plots**

```python
plt.plot(s,np.cos(s))
plt.title("cosine Curve")
plt.xlabel("x")
plt.ylabel("cos(x)")
```

**Visualizing Errors**

**Errorbars**

```python
x=np.linspace(0,10,100)
dy=0.1
y=np.sin(x)+dy*np.random.randn(100)
plt.errorbar(x,y,yerr=dy,fmt='.b');
```

**Histograms**

```python
plt.style.use
data=np.random.randn(10000)
print(data)
```

```python
plt.hist(data);
```

```python
plt.hist(data,bins=30,density=True,alpha=0.6,histtype='stepfilled')
```

```python
plt.hist(data,bins=30,density=True,alpha=0.2,histtype='stepfilled');
```

```python
x=np.random.normal(0,0.8,1000)
y=np.random.normal(-2,1,1000)
kwargs=dict(histtype='stepfilled',alpha=0.5,density=True,bins=50)
plt.hist(x,**kwargs)
plt.hist(y,**kwargs)
```

```python
counts,bin_edges=np.histogram(data,bins=5)
print(counts)
```

**2-D Histograms**

```python
mean=[5,2]
cov=[[3,1],[2,0.5]]
x,y=np.random.multivariate_normal(mean,cov,10000).T
```

**plt.hist2d:2-D Histogram**

```python
plt.hist2d(x,y,bins=20,cmap='Blues')
cb=plt.colorbar()
```

**plt.hexbin:Hexagonal binnings**

```python
plt.hexbin(x,y,gridsize=20,cmap='Greens')
cb=plt.colorbar()
```
